* 5 :: Create function to create csv_file to save data in it

```python
def get_employees_data(url):
    with open("Employees_data.csv" , "w") as f :
        writer = csv.DictWriter( f , fieldnames= [ "ID" , "Name" , "E_mail" , "Gender" , "Status" , "creation" , "Updation"])
        writer.writeheader()
        res = R.get(url)
        res_json = json.loads(res.text)
        Employees = res_json["data"]
        for emp in Employees :
            Emp_Id     =emp["id"]
            Emp_Name   =emp["name"]
            Emp_e_mail =emp["email"]
            Emp_Gender =emp["gender"]
            Emp_Status = emp["status"]
            Emp_created_at =emp["created_at"]
            Emp_updated_at =emp["updated_at"]
            writer.writerow({"ID":Emp_Id ,"Name" : Emp_Name ,"E_mail" :Emp_e_mail ,"Gender" :Emp_Gender , "Status": Emp_Status, "creation":Emp_created_at ,"Updation" :Emp_updated_at})
            print(f"U_Id :: {Emp_Id}\nEmp_Name :: {Emp_Name}\nEmp_e_mail :: {Emp_e_mail}\nEmp_Gender :: {Emp_Gender}\nEmp_Status :: {Emp_Status}\nEmp_created_at:: {Emp_created_at}\nEmp_updated_at :: {Emp_updated_at}\n-------------------")
            
    
```
