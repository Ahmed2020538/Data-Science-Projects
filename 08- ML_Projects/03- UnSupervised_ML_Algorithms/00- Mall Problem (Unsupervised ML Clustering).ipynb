```python
# Custimize autopct valeue
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:1.0f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

<p id = "Prepration">Data-Prepration</p>

```python
# preprate input data ::
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

* Data-Normalization ::
**************************
* Its one of the ost efficient phase on data scientist life-cycle which calling data in specific scaler Becouse data have different scaller.

```python
# Scaling data use standerscaler ::
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x = scaler.fit_transform(x)
```

<p id = "Model">Bulid_ML_Model</p>

* Import Requried Clusterig Algorithms ::
    * KMean-Cluster.
    * Agglomerative-Cluster.

```python
from sklearn.cluster import KMeans , AgglomerativeClustering
from scipy.cluster import hierarchy as sch
```

* Bulid KMean_Cluster Model ::

```python
# At first select random num of cluster ::
Model1 = KMeans(n_clusters=3)
y_labels = Model1.fit_predict(x)
y_labels
```

```python
# Predict new value cluster
Model1.predict([[20, 88]])
```

```python
# get Cluster_centers
centers = Model1.cluster_centers_
centers
```

```python
# Present inertia value
Model1.inertia_
```

```python
# Present data after Normalization it
x = pd.DataFrame(data=x, columns=['Annual Income (k$)', 'Spending Score (1-100)'])
x
```

```python
# visualising the clusters
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=y_labels, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], s=100, c='red')
```

<p id = "Evaluation">Evaluate_Model</p>

```python
# Apply Elbow-method to choose the best number of clusters (K Value) using Elbow Method
scores = []
for i in range(1, 11):
    Model1 = KMeans(n_clusters=i)
    Model1.fit(x)
    scores.append(Model1.inertia_)
plt.plot(range(1, 11), scores)
plt.title("Elbow Method")
plt.xlabel("num of clusters")
plt.ylabel("Score")
```

```python
# from the scores, turns out the best clusters number is 5
Model1 = KMeans(n_clusters=5)
y_labels = Model1.fit_predict(x)
y_labels
```

```python
# visualising the correct numbers of clusters
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'], c=y_labels, cmap='viridis')
centers = Model1.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], s=100, c='red')
```

* Bulid Hirarical_Cluster Model ::

```python
# Present dendrogram to choice the correct number of clusters ::
sns.set(rc = {"figure.figsize" : [60 , 40]} , font_scale = 6.7)
plt.figure(figsize=[60,40])
dndro_gram = sch.dendrogram(sch.linkage(x , method = "ward"))
plt.title("The Dndrogram of data")
plt.xlabel("Clusters")
plt.ylabel("euclidean")
plt.xticks(rotation = 90)
plt.show()
```

```python
# Build Model ::
Aglom_Model = AgglomerativeClustering(n_clusters=5 , affinity="euclidean" , linkage="ward")
Aglom_Model.fit(x)
```

```python
y_hc = Aglom_Model.fit_predict(x)
```

```python
y_hc[:50]
```

```python
# visualising the correct numbers of clusters
sns.set(rc = {"figure.figsize" : [20 , 10 ]}, font_scale=1.9)
plt.scatter(x['Annual Income (k$)'], x['Spending Score (1-100)'] , c = y_hc , cmap = "viridis")
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```
