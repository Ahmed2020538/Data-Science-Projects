```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans , AgglomerativeClustering
from scipy.cluster import hierarchy as sch
```

```python
Model1 = KMeans(n_clusters=2)
y_labels = Model1.fit_predict(X)
y_labels
```

```python
centers = Model1.cluster_centers_
centers
```

```python
Model1.inertia_
```

```python
# choose the best number of clusters (K Value) using Elbow Method
scores = []
for i in range(1, 11):
    Model1 = KMeans(n_clusters=i)
    Model1.fit(X)
    scores.append(Model1.inertia_)
plt.plot(range(1, 11), scores)
plt.title("Elbow Method")
plt.xlabel("num of clusters")
plt.ylabel("Score")
```

```python
# from the scores, turns out the best clusters number is 5
Model1 = KMeans(n_clusters=2)
y_labels = Model1.fit_predict(X)
y_labels
```

```python
df["Cluster"]=df["Private"].apply(lambda x : 1 if df["Private"] == "Yes" else 0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(df["Cluster"],Model1.labels_))
print(4*"\n")
print(classification_report(df["Cluster"],Model1.labels_))
```

```python
Aglom_Model = AgglomerativeClustering(n_clusters=2 , affinity="euclidean" , linkage="ward")
Aglom_Model.fit(X)
```

```python
sns.set(rc = {"figure.figsize" : [60 , 40]} , font_scale = 6.7)
plt.figure(figsize=[60,40])
dndro_gram = sch.dendrogram(sch.linkage(X , method = "ward"))
plt.title("The Dndrogram of data")
plt.xlabel("Clusters")
plt.ylabel("euclidean")
plt.xticks(rotation = 90)
plt.show()
```
