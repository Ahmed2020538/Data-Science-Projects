```python
Cat_cols = df.select_dtypes(include=["object"])
Num_cols = df.select_dtypes(include=["number"])
for col in Cat_cols :
    df[col].fillna(df[col].mode()[0] , inplace = True)
    df[col] = df[col].apply(lambda x : df[col].mode()[0] if x == 0 else x)
for coll in Num_cols :
    df[coll].fillna(df[coll].median() , inplace=True)
    df[coll] = df[coll].apply(lambda x : df[coll].mean()if x == 0 else x)
    
```

```python
# Create fun to find info of  the lowest and highst index in profit column

def find_min_max(x) :
    
    # using "idxmin" to find lowes index and using "idxmax" fo find highst index 
    low_index = df[x].idxmin()
    high_index = df[x].idxmax()
    
    # create 2 df of lowest  and highst indexs
    low = pd.DataFrame(df.loc[low_index , :])
    high = pd.DataFrame(df.loc[high_index , :])
    
    # maping low_index and high_index with data["name"]
    print(f"The Lowest {x} Move :: Which its index :: {low_index} ::  Its Title ::  {df['original_title'][low_index]}")
    print(f"The highst {x} Move :: Which its index :: {high_index}  :: Its Title ::   {df['original_title'][high_index]}")
    
    # Merge 2 df together as columns  
          
    return pd.concat([low , high] , axis = 1 , keys = ["Lowest" , "Highest"])
          
    
    
```

```python
find_min_max("Profit")
```

* From the result above we can see ::
     * 1 _ The first column shows the highest profit made by a movie and second column shows the highest in loss movie in this dataset
     * 2 _ As we can see that 'Avatar' movie Directed by James Cameron earn the highest profit in all, making over 2.5B in profit in this dataset.And the most in loss movie in this dataset is The Warrior's Way. Going in loss by more than 400M was directed by Singmoo Lee.

```python
# Create fun to show the visualize of output column

def visual_plot(x ,column , title , xlabel ) :
    
    # Sorting X decending order and store it in new dataframe
    info = pd.DataFrame(df[x].sort_values(ascending = False))
    
    # craate new colums in info df which have the same names of data movie name
    info["original_title"] = df["original_title"]
    
    # Mapping the data on info names
    dt = list(map(str , (info["original_title"])))
    x = list(dt[:10])
    y = list(info[column][:10])
    
    #make a plot usinf pointplot for top 10 profitable movies.
    fig = sns.pointplot( x = y , y = x)
    
    #setup the title and labels of the plot.
    fig.set_title(f"The Top  10 Movies ::{title}" , fontsize = 20)
    fig.set_xlabel(xlabel , fontsize = 13)
    
    #set the style sheet
    sns.set_style("whitegrid")
    
    
```
